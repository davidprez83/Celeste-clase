```python
import numpy as np
import matplotlib.pyplot as plt
import pymcel as pym 
import scipy.integrate as integrate
```

```python
# Posición del Sol en el sistema de referencia del baricentro del sistema solar (SSB) hoy

tabla, jd, X_sol = pym.consulta_horizons(id = 'Sun', 
                                    location = '@SSB',
                                    epochs = '2026-02-26 00:00:00',
)

# Posición de la Tierra en el sistema de referencia del baricentro del sistema solar (SSB) hoy
_,_, X_tierra = pym.consulta_horizons(id = '3',
                                    location = '@SSB',
                                    epochs = '2026-02-26 00:00:00',
)
# Posición de jupiter hoy 
_,_, X_jupiter = pym.consulta_horizons(id = '5',
                                    location = '@SSB',
                                    epochs = '2026-02-26 00:00:00',
)
```

```python
# Velocidades y posiciones

r_sol = X_sol[:3]
r_tierra = X_tierra[3:]
r_jupiter = X_jupiter[:3]

v_sol = X_sol[3:]
v_tierra = X_tierra[3:]
v_jupiter = X_jupiter[3:]
```

```python
# Vector de estado mañana

# sol 
_, _, X_sol_manana = pym.consulta_horizons(id = 'Sun',
                                    location = '@SSB',  
                                    epochs = '2026-02-27 00:00:00',
)

# tierra
_, _, X_tierra_manana = pym.consulta_horizons(id = '3',
                                              location = '@SSB',
                                              epochs = '2026-02-27 00:00:00',
)

# jupiter
_, _, X_jupiter_manana = pym.consulta_horizons(id = '5',
                                              location = '@SSB',
                                              epochs = '2026-02-27 00:00:00',
)

```

```python
# posiciones y velocidades mañana 

r_sol_manana = X_sol_manana[:3]
r_tierra_manana = X_tierra_manana[:3]
r_jupiter_manana = X_jupiter_manana[:3]

v_sol_manana = X_sol_manana[3:]
v_tierra_manana = X_tierra_manana[3:]
v_jupiter_manana = X_jupiter_manana[3:]

```

```python
solmu = pym.constantes.mu_sun
tierramu = pym.constantes.mu_earth
jupitermu = pym.constantes.mu_jupiter
```

```python
rjup_sol = r_jupiter - r_sol
rjup_tierra = r_jupiter - r_tierra

rtierra_sol = r_tierra - r_sol
rtierra_jupiter = r_tierra - r_jupiter

rsol_tierra = r_sol - r_tierra
rsol_jup = r_sol - r_jupiter
```

```python
# velocidad de jupiter
vj_mananasim = v_jupiter - (jupitermu * rjup_sol / np.linalg.norm(rjup_sol)**3 + jupitermu * rjup_tierra / np.linalg.norm(rjup_tierra)**3) 

# velocidad de la tierra 
vt_manana_sim = v_tierra - (tierramu * rtierra_sol / np.linalg.norm(rtierra_sol)**3 + tierramu * rtierra_jupiter / np.linalg.norm(rtierra_jupiter)**3) 

# velocidad del sol
vs_manana_sim = v_sol - (solmu * rsol_tierra / np.linalg.norm(rsol_tierra)**3 + solmu * rsol_jup / np.linalg.norm(rsol_jup)**3)
```
